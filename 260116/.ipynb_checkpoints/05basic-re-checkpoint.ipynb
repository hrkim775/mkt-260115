```python
# 사용자에서 3개의 숫자를 입력받은 후 그 중 가장 작은 숫자를 출력하세요

"""
number = int(input("3개의 숫자를 입력하세요"))

if number 

number.reverse()
"""


num01 = int(input("첫번째 숫자"))
num02 = int(input("두번째 숫자"))
num03 = int(input("세번째 숫자"))

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
elif num03 <= num01 and num03 <= num02 :
    print(num03)
```

```python
# 위에 있는 셀의 문제와 동일한데 좀 더 실용적인 코드

numbers = []

for i in range(3) : 
    number = int(input(f"{i + 1}번째 숫자 : "))
    numbers.append(number)

# print(numbers)


#구조분해할당 문법 : 리스트와 같이 여러개 변수값을 가지고 있는 자료구조의 값들을 하나씩 추출 > 독립변수로 담을 수 있는 문법

num01, num02, num03 = numbers

if num01 <= num02 and num01 <= num03 :
    print(num01)
elif num02 <= num01 and num02 <= num03 :
    print(num02)
elif num03 <= num01 and num03 <= num02 :
    print(num03)
```

```python
# 위에 있는 셀의 문제와 동일한데 좀 더더 실용적인 코드

numbers = []

for i in range(3) : #반복(문 : 식이 여러개가 있다 / 식 + 식 / 여러가지 식과 구문이 합쳐져 있는 문장이다)
    number = int(input(f"{i + 1}번째 숫자 : "))
    numbers.append(number)

num01, num02, num03 = numbers

print(min(numbers))
```

```python
# 위에 있는 셀의 문제와 동일한데 좀 더더더 실용적인 코드 (현업에서 사용하는 코드)
# 리스트 컨프리핸션 = list comprihension

nums = [int(input(f"{i + 1}번째 숫자 : ")) for i in range(3)]

print(min(nums))


# 오름차순 함수를 사용한다면?

numbers = [int(input(f"{i + 1}번째 숫자 : ")) for i in range(3)] # 리스트의 형태를 띄고 있지만, 실제 리스트의 속성값을 모두 사용할수 없음
print(numbers)
print(sorted(numbers)[0])

# 왜 sort 는 안되고 sorted 는 되냐?
# sort는 원래 값을 편집하기 때문 -> 인덱스의 구조 X
# 반면 sorted는 완전 새로운 리스트 구조 -> 인덱스의 구조 O
```
