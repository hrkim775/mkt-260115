```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By 
from webdriver_manager.chrome import ChromeDriverManager
from urllib import parse
import time
import pandas as pd # as는 별명을 부여한다는 뜻

service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless") # 가상의 브라우저를 우리 눈에 보이지 않도록 설정
# options.add.argument("--desable-gpu")
options.add_argument("--disable-dev-shm-usege")
options.add_argument("--window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36")
options.add_argument("--lean=ko_KE")


driver = webdriver.Chrome(service=service, options = options)

query = "퍼포먼스마케터"
encoded_query = parse.quote(query)

max_page = 6

results = []


# 슬라이스와 동일하게 +1을 해줘야 원하는 영역까지 나옴
# for문을 쓸때는 반드시!!!! for 보다 들여쓰기가 되어있어야한다
for page in range(1, max_page + 1) : 
    url = f"https://www.jobkorea.co.kr/Search?stext={encoded_query}&Page_No={page}"
    driver.get(url)
    time.sleep(1)

    job_posts = driver.find_elements(By.CSS_SELECTOR, "div.styles_mb_space2__dk46ts4t")
    
# 중첩for문 / for문 안에 또 다른 for문이 있다
    for post in job_posts :
        title_elem = post.find_element(By.CSS_SELECTOR, "a.Flex_display_flex__i0l0hl2.Flex_gap_space6__i0l0hl1g.Flex_align_center__i0l0hl8.styles_mb_space2__dk46ts4t")
        title = title_elem.text.strip()
        link = title_elem.get_attribute("href")
        results.append({
            "공고제목": title
            "지원링크": link
        })
        
        
        # try : 
        #     title_elem = post.find_element(By.CSS_SELETOR, "a.Flex_display_flex__i0l0hl2.Flex_gap_space6__i0l0hl1g.Flex_align_center__i0l0hl8.styles_mb_space2__dk46ts4t")
        #     print(title_elem.text.strip())
        # except : 
        #     continue

driver.quit()

df = pd.DataFrame(results)
df.to_excel("잡코리아_퍼포먼스마케터.xlsx", index=False)
print(("잡코리아_퍼포먼스마케터.xlsx 저장완료")
    
   # "https://www.jobkorea.co.kr/Search?stext=+ encoded_query + "&Page_No=" + page
```
