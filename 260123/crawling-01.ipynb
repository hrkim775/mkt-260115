```python
# HTTP (hyper Text Tranfer Protocol ) 상태코드

# 1xx (정보): 요청을 받았으며 프로세스를 계속한다
# 2xx (성공): 요청을 성공적으로 받았으며 인식했고 수용하였다
# 3xx (리다이렉션): 요청 완료를 위해 추가 작업 조치가 필요하다
# 4xx (클라이언트 오류): 요청의 문법이 잘못되었거나 요청을 처리할 수 없다 -> 가장 많이 보게된다
# 5xx (서버 오류): 서버가 명백히 유효한 요청에 대해 충족을 실패했다

# 서버측에서 응답 메시지를 의도적으로 차단하거나 오류를 발생시키는 경우도 있음
# 이런 경우 예외조항처리
#크롤링을 하려고 했는데 정상적으로 못했을때 -> 이유가 뭔지 알 수 없을때 -> 예외조항처리 해볼 수 있다

from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
if response.status_code != 200 :
    print("페이지 없음")
else :
    soup = BeautifulSoup(response.content, "html.parser")
    data = soup.select(".card-tex")
    for item in data :
        print(item.get_text(" ", strip=True))



# soup = BeautifulSoup(respose.content, "html.parser")  
```

```python
# 복수의 다중페이지 크롤링 하는 방법
# 1) 각 페이지별 주소창에 주소값을 확인한다
# 2) 페이지별 주소값에 차이점 즉, 패턴을 확인한다
# 3) 확인한 패턴에 부합하는 파이썬 코드를 작성해서 페이지별 크롤링을 진행한다

# https://davelee-fun.github.io/index.html
# https://davelee-fun.github.io/page2/
# https://davelee-fun.github.io/page3/

# https : 프로토콜 (통신규약)
# davelee-fun.github.io : 도메인 주소
# /page2/ : 파라미터 -> 경로를 정의하는 것 (2페이냐, 3페이지냐..)
# index.html : 기본 페이지 -> 이건 경로랑 상관없다. 없어도 된다

import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code != 200 :   # 예외조항처리
    print("페이지 없음")
else :
    # 8회의 반복 -> 0 ~ 7 까지 8회 반복

    for page_num in range(8) :
        if page_num == 0 :
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else :
            response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
    
        soup = BeautifulSoup(response.content, "html.parser")
        data = soup.select(".card-text")
        for item in data :
            print(item.get_text(" ", strip=True))
```

```python
import openpyxl

# 데이터 크롤링 (수집)
# 데이터 저장 준비 단계
# 엑셀에 저장 (저장) 
# 1개의 코드에서 모두 진행하는 것은 좋지 않다
# 리팩토링 (refactiring) : 중요한 데이터를 따로 따로 작성하는 것

def write_excel_template(filename, sheetname, listdata) :
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions["A"].width = 100  #column_dimensions : 엑셀의 열

    if sheetname != "" :
        excel_sheet.title = sheetname

    for item in listdata : 
        excel_sheet.append(item)    # 자료가 한 줄 (세로행)로 들어감

    excel_file.save(filename)   # 저장할 파일의 이름
    excel_file.close()    # 종결
    
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code != 200 :
    print("페이지 없음")
else :

    product_list = list()
    # product_list = []
    
    for page_num in range(8) :
        if page_num == 0 :
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else :
            response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
    
        soup = BeautifulSoup(response.content, "html.parser")
        data = soup.select(".card")

        for item in data :
            product_name = item.select_one(".card-text")
            product_date = item.select_one(".post-date")
            product_info = [product_name.get_text(" ", strip=True), product_date.get_text(" ", strip=True)]
            product_list.append(product_info)
```
