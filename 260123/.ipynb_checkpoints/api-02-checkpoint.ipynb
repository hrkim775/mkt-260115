```python
import requests

url = "https://apis.data.go.kr/B551011/GoCamping/basedList"

# 우리가 크롤링을 할때는 일반 웹브라우저인 척을 하려고 headers를 썼던 것이고 지금은 그럴 필요가 없음

request_params = {
    "numOfRows":"10",
    "pageNo":"1",
    "MobileOS":"ETC",
    "MobileApp":"camping",
    "serviceKey":"32852873e25d85bf199a671eb5c37d91ec4ac0ae04f9770f5a1bc2a5a249ee87",
    "_type":"xml"
}

response = requests.get(url, params=request_params)
print(response)
```

```python
import subprocess
import xml.etree.ElementTree as ET

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=32852873e25d85bf199a671eb5c37d91ec4ac0ae04f9770f5a1bc2a5a249ee87&_type=xml"

result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
root = ET.fromstring(result.stdout) 

print(root.findall(".//item"))
```
