```python
# 반복문 : 
# -사람이 매번 동일한 혹은 유사한 행위를 반복하다보면 -> 실수, 누락이 생김
# - 속도도 기계보다 빠를 수 없음
# - AI, 계산기, 알파고 등이 나올 수 있는 이유는 반볻기능 때문이다 (빠르고 실수가 없음)
# - 다른 프로그램에 비해 파이썬의 반복문은 단촐
# - for, while 두 가지만 기억
# 반복 혹은 조건문을 사용할때는 반드시 실행문이 들여쓰기 되어있어야 한다

"""

for 변수 in (반복할 대상/개체 = iter한 속성을 가지고 있어야함 = 리스트 | 문자열) :
    실행문

# 반복하는 대상이 명확하지 않을때 range 사용
for 변수 in range(인수 | 인자값) :
    실행문 -> 반드시 0부터 시작함

for 변수 in range(첫번째 인수 = 어떤 숫자로 시작할 것인가, 두번째 인수 = 해당 숫자에서 -1 값) :
    실행문 

    
"""
```

```python
for index in range(10) :
    print(index)
```

```python
for index in range(1, 11) :
    print(index)
```

```python
sum = 0

for index in range(1, 11) : # 첫번째 인자값 ~ 두번째 인자값 -1까지 즉, 1 ~ 10까지 / 슬라이싱과 유사
    sum = sum + index # 누산기

print(sum)
```
