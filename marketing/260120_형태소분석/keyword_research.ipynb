```python
from openpyxl import load_workbook
from kiwipiepy import Kiwi

src = r"naver keyword.xlsx"
out = r"naver keyword_result.xlsx"

wb = load_workbook(src)
ws = wb.active

def find_col_by_header(ws, keywords, header_row = 1) : # keyword =  ["연관키워드", "월간검색수 (pc)", "원간검색수 (mobile)", ..., "월평균 노출 광고수"]
    for col in range(1, ws.max_column + 1) :
        v = ws.cell(row=header_row, column=col).value
        if not v :
            continue
        s = str(v).replace(" ",'') # 월간검색수 (PC) - > 월간검색수(PC) 이렇게 빈 칸 수정하는 구문

        ok = True
        for kw in keywords :
            if kw.replace(" ", '') not in s :
                ok = False
                break
            if ok :
                return col
        return None

        

print(ws)
```
