# git 사용방법
1) 로컬컴퓨터 (내 컴퓨터) : git 응용프로그램 설치
2) 터미널 : git -v 명령어
- 검색바 > cmd > 클릭 : 검정색
3) 만약, 없다면 > 구글 크롬 > git 다운로드 후 설치
- chocholatey : 대리로 응용프로그램 다운로드 후 설치해주는 중간 프로그램 -> 윈도우용
4) 깃의 소유주 확인 & 변경
- git config --list
5) github 회원가입 : 계정정보 값으로 입력 : https://github.com/
- git : 나의 로컬컴퓨터에서 버전 관리 툴
- github : 나의 로컬컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할 떄, 혹은 내가 사용하고 있는 디바이스가 2대 이상인 경우, A에서도 데이터를 관리, B에서도 데이터관리
- repository : github > 프로젝트에 따른 개별 폴더
- user.email : hrkim775@gmail.com
- user.name : hrkim775
- 터미널 재접속 > 관리자모드 : 전역(global)세팅
- git config --global user.name "hrkim775"
- git config --global user.email "hrkim775@gmail.com"
6) git commit & push
- cd "현재 나의 루트폴더 " 이동 (내 파일들을 저장해놓은 폴더 경로를 의미)
- git 초기화 : git init (정상적으로 작동 되었다면 저장해놓은 폴더 안에 .git 폴더가 생성되어있어야함 / 안보인다면 숨김파일보기 해보기)
- staging
