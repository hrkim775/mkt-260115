### 문자열 관련 파이썬 기초 문법

1. 문자열에 있는 특정 문자 갯수 세기 : count()
2. 문자열에 있는 특정 문자의 위치 파악 : index() -> 반드시 있어야 하는 값을 찾을때만 찾아와야한다 / 아니면 에러가 난다
3. 문자열에 있는 특정 문자의 위치 파악 (에러 발생 X) : find()
4. 서로 다른 문자열을 하나로 합치고자 할 때 : join()
5. 문자열의 앞/뒤 공백 없앨 때 : strip() // 오른쪽에 여백을 없애기 = rstrip() // 왼쪽에 여백을 없애기 = lstrip() -> 인자값이 없으면 기본적으로 공백을 없앰
6. 대/소문자 변경 : 대문자화 = upper() // 소문자화 = lower()
7. 문자열을 쪼개고자 할 때 : split() -> 리스트를 반환 -> 인자값이 없으면 기본적으로 공백을 기준으로 나눈다
8. 특정 문자열을 찾아서 원하는 문자열로 대체하고자 할 때 : replace(교체할 대상, 무엇으로 교체할지)

```python
# 문자열에 있는 특정 문자 갯수 세기 : count()

data = "Dave David"
data.count("D")
```

```python
# 문자열에 있는 특정 문자의 위치 파악 : index() -> 반드시 있어야 하는 값을 찾을때만 찾아와야한다 / 아니면 에러가 난다

string = "Dave Id is Dave"
string.index("x")
```

```python
#문자열에 있는 특정 문자의 위치 파악 (에러 발생 X) : find()

string = "Dave Id is Dave"
string.find("x")
```

```python
# 서로 다른 문자열을 하나로 합치고자 할 때 : join()

string = "12345"
comma = ","

comma.join(string)
",".join(string)  # 위에 문장이랑 똑같은 결과 도출함 / 사이사이에 넣는 것
```

```python
# 문자열의 앞/뒤 공백 없앨 때 : strip()

data = "    Dave "   # 공백이 아무리 많아도 다 없앰
data.strip()
```

```python
data = "    Dave "   # 오른쪽
data.rstrip()
```

```python
data = "    Dave "   # 왼쪽
data.lstrip()
```

```python
# 없애고 싶은 요소 없앨때고 사용

string = "     999(DAVE)8888888  "
string.strip("98() ")
```

```python
# 대/소문자 변경 : 대문자화 = upper() // 소문자화 = lower()

string = "Dave"
# string.upper()
string.lower()
```

```python
string = "Dave goes to Korea"
string.split()[3]   # 구분자를 아무것도 넣지 않는다면 , 공백을 구분자로 선택한다
```

```python
string = "Dave/goes to/Korea"
string.split("/")   # 안에 있는 구분자에 따라 쪼개는 기준이 달라진다
```

```python
string2 = "(David)"
string2.replace("(", "").replace(")", "")   # 이런 문법을 method chaining 기법이라고 한다
```
