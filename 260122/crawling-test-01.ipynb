```python
import requests
from bs4 import BeautifulSoup

url = "https://crawlingtest-260122.netlify.app/"
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}


response = requests.get(url, headers=headers)
response.encoding = "utf-8"

soup = BeautifulSoup(response.text, "html.parser")

# 상품명은 여럭개가 존재
# 복수의 상품명은 공통적인 특징 : product-name
# 한번에 크롤링해서 동시에 3개를 찾아올 수 있어야한다
# 리스트 형태로 가져올 수 있어야한다

product = soup.select("h2.product-name")

for products in product : 

    title = products.text
    print(title)


names = [tag.text.strip() for tag in soup.select("h2.product-name")]     # 리스트 컴프리핸션
print(names) 


# 상품명, 카테고리명

for card in soup.select(".product-card") : 
    name = card.select_one(".product-name").text.strip()
    category = card.select_one(".category").text.strip()
    
    print(f"{name} / {category}")



# 상품 당 가격 출력

for card in soup.select(".product-card") : 
    name = card.select_one(".product-name").text.strip()
    #price = card.select_one(".price").text.strip()   # 이렇게하면 숫자가 문자로 나옴

    price = int(card.select_one(".price")["data-price"])   # 명시적형변환
    print(type(price))
    
    print(f"{name} / {price:,}")  # :, 일시적으로 보기 편하게 숫자에 , 찍어주는 것



#new, hot 뱃지가 있는 상품만 상품명을 크롤링해오기

for card in soup.select(".product-card") : 
    badge = card.select_one(".badge")

    if badge :
        name = card.select_one(".product-name").text.strip()
        print(f"{name} / {badge.text.strip()}")




# 5만원 이상의 상품의 이름만 크롤링 해오기

for card in soup.select(".product-card") : 
    price = int(card.select_one(".price")["data-price"])

    if price >= 50000 :
        name = card.select_one(".product-name").text.strip()
        print(f"{name} / {price:,}")




# 특정 사이트에 접속해서 데이터를 수집만 했는데, 이를 다시 활용할 수 있어야한다 (프로토타입)
# 수집한 상품별로 상품명, 카테고리, 가격을 하나의 딕셔러니로 만들어주시고, 해당 딕셔너리가 하나의 리스타 안에 저장될 수 있도록 해주세요

"""
[
    {
        "name": "..."
        "category": "..."
        "price": "..."
    },
    {},
    {}

]
"""
items = []

for card in soup.select(".product-card") : 
    
    items.append({
        "name":card.select_one(".product-name").text.strip(),
        "category": card.select_one(".category").text.strip(),
        "price": int(card.select_one(".price")["data-price"])
    })
    
    print(items)
```

# requests
- select : (선택자를 기준) 복수의 값을 찾아올 때, 유사리스트 형태로 값을 변환 -> 반복문 사용가능

- text : 크롤링해온 값에서 문자만 남기고 싶을 때 사용
- strip() : 해당 문자 앞, 뒤 빈여행 혹은 이스케이프 시퀀스 제거하고자 할 때 사용
