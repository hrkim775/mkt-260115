# 파이썬 함수 (function = 기능)
- 함수 : 어떤 입력값을 받아서 내부에서 연산처리 후 해당 결과 값을 외부로 출력할 수 있도록 기능 하는 것 / 프로그래밍 언어의 꽃
- 함수의 값을 입력 & 출력 : input & output
- 함수가 생성(*선언)되어야 함수가 존재할 수 있고 그래야 값(인자값 or 인수 -> 매개변수라고도 부른다)을 입력할 수 있다
- 인자값은 넣을 수도 있고, 없을 수도 있다!
- 함수를 사용(*호출)
- 함수를 선언할땐 def가 와야한다
- 예를 들면)

  def func() :
      실제 함수의 실행연산이 진행될 공간 / *** 조건문처럼 실행문은 반드시 함수보다 들여쓰기 되어있어야한다

- 함수의 선언부 & 호출부의 결과값 "인자값/인수"의 갯수는 기본적으로 같아야한다
  어느 특정 한쪽이 상대적으로 갯수가 많거나 적으면 정상적으로 실행되지 않는다
- return(문) : 함수의 내부 연산 처리 후 특정 값을 함수 바깥으로 값을 보낼 수 있는 기능 

# 파이썬 자료구조
- 문자열 = "rtring"
- 정수 = 10
- 실수 = 3.14
- 논리 = T / F

  > 여기까지는 단일값을 저장하는 자료 형태 즉, 1:1

  
- 리스트 = [1, 2, 3] ->  리스트는 원하는 값을 추가하거나 삭제 할 수 있다
- 튜플 = (1, 2)
  리스트와 거의 같다 -> 튜플은 최초에 선언된 형태의 자료구조를 변경할 수 없다는 것이 리스트와의 차이점
  데이터의 추가 혹은 삭제 불가능 !!!!
  >iter & 반복문
  >리스트 & 튜플 안에 담겨 있는 값들이 구체적으로 어떤 값을 의미하는 지 알 수 없음
  
- 딕셔너리(**중요) = 사전 = {"apple": "사과"} -> apple은 key 라고 지칭하고, 사과는 value라고 지칭한다 / 이런 것을 속성(property) 이라고 한다
- 집합 = 그룹
  교집합, 합집합, 차집합, 공집합 등
  그룹간에 더하기, 빼기, 공통값 등을 찾을 수 있다
  반복되는 값을 허용하지 않는다
  실무에선 어떤 데이터 값을 크롤링해서 값을 찾아올 때 반복적인 값이 들어오는 경우에 이 값을 1회 처리 할때 사용한다
  
  
 > 여기까지는 다중값을 저장하는 자료형태 즉, 1:3 / 1:7


****** [] -> 리스트 , () -> 튜플, {} -> 딕셔러니


1. 리스트 선언 할때 : 변수명 = [값]
>변수명 = list()

2. 튜플 선언 할때 : 변수명 = (값)
>변수명 = tuple()


3. 딕셔너리 선언 할때 : 변수명 = {"a" : "b"}
>변수명 = dict()


4. 세트 선언 할때 : 변수명 = {}
>변수명 = set()

```python
# 함수의 선언부 & 호출부의 결과값 "인자값/인수"의 갯수는 기본적으로 같아야한다
# 어느 특정 한쪽이 상대적으로 갯수가 많거나 적으면 정상적으로 실행되지 않는다

def func3(data1, data2) :
    print(data1 + data2)
```

```python
# return() : 함수의 내부 연산 처리 후 특정 값을 함수 바깥으로 값을 보낼 수 있는 기능 

def func4(data1, data2) :
    return data1 + data2
```

```python
result = func4(1, 2)
result
```

```python
# 실제 실무에서 튜플을 사용하면 좋은 상황은 '절대로 값이 변경되면 안되는 경우'이다 예를 들어 주민등록번호 같이 고정값을 츄지해야하는 경우

tuple1 = (1, 2, 3, 4)
list1 = [1, 2, 3, 4]

print(tuple1, list1)
print(dir(tuple1))
print(tuple1[0])
```

```python
tuple1 = (1, 2, 3, 4)
tuple2 = (5, 6)

print(tuple1 + tuple2) # 튜플끼리 더하기 연산 가능하다
print(tuple1 * 3) # 튜플 내에서 값을 반복하는 기능이 가능하다

for item in tuple1 : 
    print(item * 3)
```

```python
def test_func1(a, b) : 
    quot = a // b    #  몫
    rem = a % b      # 나머지
    return (quot, rem)
```

```python
result = test_func1(10, 3)

# 구조분해할당 기능이 있다

(f, s) = test_func1(10, 3) # () = test_func1(10, 3) -> 이 순서 지켜서 써야한다

#(f, s) = (3, 1)
print(f, s)
```

```python
data1 = {"a", "b", "c"}
data2 = {"b", "c", "d"}
```

```python
# 교집합 : 두 그룹의 공통점 찾기
data1 & data2
```

```python
# 합집합 : 두 그룹의 데이터를 합한다 / 중복값은 한번만 가져온다
data1 | data2
```

```python
# 차집합 : 한 그룹에서 나머지 그룹의 값을 제저 한 값을 가져온다
data1 - data2
```

```python
# 부분집합 : 교집합을 제외한 나머지 부분 집합만 가져온다
data1 ^ data2
```

```python
# set 자료구조는 중복 값을 혀용하지 않는다
data_list = {"a", "b", "c", "d", "d", "a", "b", "b", "c"}
```

```python
data_list
```

```python
set_result = (set(data_list)) 
print (set_result )#명시적 형변환 vs 자동 형변환
```

```python
# 특정 사이트에서 데이터를 크롤링하면 대부분 중복값이 함께 들어온다
# 그럴때 세트 자료구조를 사용한다. 그걸 다시 리스트로 바꿔줘야한다

list(set_result)
```
