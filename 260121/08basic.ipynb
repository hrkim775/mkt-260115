```python
price = "100 달러"
exchange_rate = 1480

# price 라는 변수에 있는 100달러의 값을 환율을 적용해서 한화로 변환하여 값을 출력하시오

# exchange = {"1 달러": "1480"}

# 문자열 : 인덱스 -> 슬라이싱
print(type(int(price[0:3])))

result = (int(price[0:3])) * exchange_rate
print(result)
print(f"{result:,}") # 천의 단위마다 쉼표를 찍어라
```

```python
# 반복문을 활용해서 구구단 2~9단까지 출력해주세요

# for in range(9) : 

# 2 ~ 9 반복
for index_1 in range(2, 10) :
    print(f"===구구단 {index_1}단 ===")
    for index_2 in range(1, 10) : 
        print(f"{index_1} x {index_2} = {index_1 * index_2}")
print("\n") # escape sequnce => 문자처럼 생겼지만 문자 이상의 고유한 문법을 가지고 있는 문자
```
