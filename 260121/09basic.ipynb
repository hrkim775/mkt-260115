```python
# 클래스를 통해서 만들어진 매서드 함수

string = "harim kim"
print(string.split())
```

```python
# 클래스를 선언

class Student : 
    def __init__(self, name, grade) :
        self.name = name
        self.grade = grade # 클래스의 이름은 마음대로 쓸수 있지만 그 외 것은 무조건 고정

    def study(self) :
        print(self.name, "공부")

    def take(self, sub) :
        print(self.name, sub, "공부")
```

```python
student1 = Student("harim", 3)
student2 = Student("jinkyu", 4)

print(student1.name, student1.grade)
print(student2.name, student2.grade)
```

```python
student1.study()
```

```python
student1.take("수학")
```
