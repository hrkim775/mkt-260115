```python
# requests.get() : 특정 url 주소에 방문해서 값을 찾아와라
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# 로컬 컴퓨터의 브라우저 드라이버와 동일한 버전으로 설치 & 이를 통핸 객체 생성
service = Service(ChromeDriverManager().install())

# 가상 브라우저가 실제처럼 보일 수 있도록 설정하는 옵션
options = Options()
options.add_argument("--headless") # 가상의 브라우저가 띄워졌음에도 우리 눈에는 안나타나도록 하는 것
options.add_argument("--disable-gpu") # 굳이 가상 브라우저를 우리 눈에 보이지 않겠다면, gpu작동이 불필요하기 때문에 --headless & --disable-gpu같이 움직인다
options.add_argument("--disable-dev-shm-usage") # 가상 메모리를 사용하지 말라 : 
options.add_argument("--window-size=1920,1080") # ,를 쓰시도 하고 x를 쓰기도 함 / 가상 브라우저의 사이즈 정보 설정
options.add_argument("--start-maximized") # 반응형 사이트의 경우 사이즈를 데스크탑(가장 큰 사이즈)에 맞추는 것
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR") # 브라우저를 사용하는 해당 국가의 모국어
options.add_argument("--no-sandbox") # sandbox는 규제를 뜻함 : 크롤링자체를 차단하는 가이드, 메뉴얼이 있을때 그걸 우회해서 들어갈 수 있다



# 브라우저 드라이버 정보값을 가지고 있는 객체를 매개변수로 받아서 크롤링 준비 완료
driver = webdriver.Chrome(service=service) # 크롬브라우저를 통해 간다
driver.get("https://davelee-fun.github.io/") # 이 사이트에 방문해서 값을 찾아와라

elements = driver.find_elements(By.CSS_SELECTOR, "div.wrapthumbnail img")

for element in elements : 
    print(element.get_attribute("src"))

    
driver.quit()
```

### selenium 문법
(find (단일 태그 중심으로 값을 찾아옴)/ find_all (복수의 태그~) / select (복수의 선택자를 기준으로 값을 찾아옴) / select_one (단일 선택자 ~))
1. find_elements : 복수의 값을 찾아온다 -> 리스트 형태 / 동일한 값을 전체 다 찾아서 가져온다
2. find_element : 단수의 값을 찾아온다 -> 리스트 형태 / 위에서부터 값을 읽다가 최초로 발견한 값을 찾아오고 종료
3. By :
   .TAG_NAME (태그 중심으로 값을 찾아온다)
   .ID (아이디 중심으로 값을 찾아온다) -> ID는 문서상 한번밖에 못 쓴다
   .CSS_SELECTOR (선택자 중심)
   .NAME (네임의 값만 찾아올 수 있다)
5. get_attribute() : 특정 속성의 속성값을 찾아오고자 할 때 사용
6. .clear() : 특정 영역 내 값을 삭제 (form 요소 기준 > input, textarea 외외)
7. .send_keys() : 어떤 값 혹은 키보드의 키값을 전달
8. Keys : 키보드의 키를 입력하고자 할 때
9. sleep() : n초의 시간동안 지연시키고자 할 때
10. screenshot(filename) : filename으로 스크린캑쳐 가느
11. set_window_size(가로, 세로) : 윈도우 창을 가로, 세로 사이즈로 조정

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time


service = Service(ChromeDriverManager().install())


options = Options()
# options.add_argument("--headless") 
# options.add_argument("--disable-gpu") 
options.add_argument("--disable-dev-shm-usage") 
options.add_argument("--window-size=1920,1080") 
options.add_argument("--start-maximized") 
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR") 
options.add_argument("--no-sandbox") 


driver = webdriver.Chrome(service=service) 
driver.get("https://davelee-fun.github.io/") 

driver.execute_script("window.scrollTo(0, document.body.scrollHeight)") # 스크롤을 내리라는 명령어. 이거말고 쓸 수 있는게 없음
# element = driver.find_element(By.CSS_SELECTOR, "div.mc-field-group > input")
# print(element.get_attribute("name"))
element = driver.find_element(By.NAME, "EMAIL")
element.clear() # placeholder의 내부 값을 제거

element.send_keys("jinkyu@gmail.com")

time.sleep(4)

element.send_keys(Keys.RETURN)

time.sleep(4)

driver.quit() # 셀레니움을 활용한 가상 브라우저의 종료
```
